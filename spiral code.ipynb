```python
def star(n):
    print(' '* (n) +"*"+' '*(n)+'*')
    return star(n-1) if n >1 else ''
    

print(star(5))
```

```python
def generate_spiral(n):
    # Create an n x n matrix filled with zeros
    matrix = [[0]*n for _ in range(n)]
    
    value = 1
    top, bottom = 0, n-1
    left, right = 0, n-1

    while value <= n*n:
        # left to right
        for i in range(left, right+1):
            matrix[top][i] = value
            value += 1
        top += 1

        # top to bottom
        for i in range(top, bottom+1):
            matrix[i][right] = value
            value += 1
        right -= 1

        # right to left
        for i in range(right, left-1, -1):
            matrix[bottom][i] = value
            value += 1
        bottom -= 1

        # bottom to top
        for i in range(bottom, top-1, -1):
            matrix[i][left] = value
            value += 1
        left += 1

    return matrix

def print_matrix(matrix):
    for row in matrix:
        print(' '.join(f"{num:2}" for num in row))

# Call function for 5x5
n = 7
spiral = generate_spiral(n)
print_matrix(spiral)

```

```python
def snake(n):
    lis=[["_"]*n for _ in range(n)]
    top, bottom = 0, n-1
    left, right = 0, n-1
    value = 1
    while value <= n*n:
            for i in range(top,right+1):
                lis[top][i] = "#"
                value += 1
                for row in lis:
                    print(' '.join(f"{num:2}" for num in row))
                print()  # Print a new line after each row

            top += 1
            for i in range(top,bottom+1):
                lis[i][right] = "#"
                value += 1
                for row in lis:
                    print(' '.join(f"{num:2}" for num in row))
                print()

            right -= 1
            for i in range(right, left-1, -1):
                lis[bottom][i] = "#"
                value += 1
                for row in lis:
                    print(' '.join(f"{num:2}" for num in row))
                print()

            bottom -= 1
            for i in range(bottom, top-1, -1):
                lis[i][left] = "#"
                value += 1
                for row in lis:
                    print(' '.join(f"{num:2}" for num in row))
                print()

            left += 1
    return lis

def print_matrix(matrix):
    for row in matrix:
        print(' '.join(f"{num:2}" for num in row))

snake_pattern = snake(5)
print_matrix(snake_pattern)
```
