1.TWO SUM PROBLEM

```python
nums = [3,2,4]
target = 6

for i in range(len(nums)):
    diff = target - nums[i]
    for j in range(i + 1, len(nums)):
        if nums[j] == diff:
            print(i, j)
```

--. NUMBER AND BASE POLINDROME == K

```python
def is_base_poli(number,base):
    temp=""
    while number>0:
        rem=number%base
        temp=str(rem)+temp
        number//=base
    if temp==temp[::-1]:
        return True
    else:
        return False
    
def is_num_poli(number,base):
    if str(number)==str(number)[::-1]:
        if is_base_poli(number,base)==True:
            return True
    
def check(k,n):
    ta=1
    lis=[]
    
    while len(lis)!=n:
        while ta>0:
            if is_num_poli(ta,k)==True and len(lis)!=n:
                lis.append(ta)
            ta+=1
            break
    sum=0
    for i in lis:
        sum+=i
    return sum

check(7,17)
```

2.SUB-ARRAY OF GIVEN ARRAY = SUM

```python

def sub(nums,start,end):
    lis=[]
    if end==len(nums):
        return lis
    elif start>end:
        return sub(nums,0,end+1)
    else:
        lis.append(nums[start:end+1])
        lis+=sub(nums,start+1,end)
        return lis
    
    
nums=[1,1,1]
k=2
count=0
for i in sub(nums,0,0):
    sum=0
    for j in i:
        sum+=j
    if sum==k:
        count+=1   
count
```

FREQUENCY OF NUMBER IN ARRAY >= LEN(NUMS)/2

```python
from collections import Counter
nums=[2,2,1,1,1,2,2]
s=Counter(nums)
for i,j in s.items():
    if j>=abs(len(nums)/2):
        print(i)
```
